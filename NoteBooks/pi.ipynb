# Understanding PI: Calculating the Circle Constant

Welcome to this notebook! Here we'll explore what PI (π) is, why it matters, and how to calculate it using basic Python. This notebook will also teach you the fundamentals of Jupyter notebooks.

## What is PI?

PI (π) is a mathematical constant that represents the ratio of a circle's circumference to its diameter.

**Key Facts:**
- π ≈ 3.14159265358979...
- It's an **irrational number** — it never repeats or terminates in decimal form
- It appears throughout mathematics, physics, and engineering
- The ancient Egyptians and Greeks discovered it by measuring circles
- Today, computers have calculated PI to trillions of digits!

## The PI Formula

The definition of PI is simple:

$$\pi = \frac{\text{circumference}}{\text{diameter}}$$

For any circle, if you measure around the edge (circumference) and divide by the width through the center (diameter), you always get approximately 3.14159...

**Example:**
- A circle with diameter 1 has circumference ≈ 3.14159
- A circle with diameter 10 has circumference ≈ 31.4159
- The ratio is always PI!

## Method 1: Geometric Approximation (Inscribed Polygon)

### The Idea
Imagine drawing a polygon (a shape with straight sides) inside a circle. The more sides the polygon has, the closer it gets to the circle's shape. If we calculate the perimeter of this polygon and know the circle's diameter, we can estimate PI.

**How it works:**
1. Start with a simple polygon (e.g., a square inscribed in a circle)
2. Calculate the perimeter of the polygon
3. Divide perimeter by diameter to estimate PI
4. Increase the number of sides (octagon, 16-gon, etc.)
5. Watch the estimate get closer to the true value of PI

### The Math
For a regular polygon with `n` sides inscribed in a circle of radius `r`:
- The perimeter approximates the circumference
- As `n` increases, perimeter → circumference
- Therefore: `perimeter / (2*r)` → PI

```python
import math

# Method 1: Inscribed Polygon Approximation
def calculate_pi_polygon(num_sides):
    """
    Estimate PI using an inscribed regular polygon.
    
    Args:
        num_sides: Number of sides of the polygon
        
    Returns:
        Estimated value of PI
    """
    # Circle with radius 1
    radius = 1
    
    # Central angle for each side of the polygon (in radians)
    angle = 2 * math.pi / num_sides
    
    # Half of the angle
    half_angle = angle / 2
    
    # Using trigonometry: half of one side length = radius * sin(half_angle)
    side_length = 2 * radius * math.sin(half_angle)
    
    # Total perimeter of the polygon
    perimeter = num_sides * side_length
    
    # PI = circumference / diameter = perimeter / (2 * radius)
    pi_estimate = perimeter / (2 * radius)
    
    return pi_estimate

# Test with increasing number of sides
print("Geometric Approximation Method (Inscribed Polygon)")
print("=" * 50)
print(f"{'Sides':<10} {'PI Estimate':<20} {'Error':<15}")
print("-" * 50)

actual_pi = math.pi

for sides in [4, 8, 16, 32, 64, 128, 256]:
    pi_est = calculate_pi_polygon(sides)
    error = abs(pi_est - actual_pi)
    print(f"{sides:<10} {pi_est:<20.10f} {error:<15.10f}")

print(f"\nActual PI: {actual_pi:.10f}")
```

## Method 2: Monte Carlo Simulation

### The Idea
Imagine throwing darts randomly at a square that contains a circle. The ratio of darts that land inside the circle versus total darts thrown can help us calculate PI!

**How it works:**
1. Generate random points in a square
2. Count how many fall inside a circle inscribed in the square
3. Use the ratio to calculate PI

**The Math:**
- Square area = (2r)² = 4r² (where r is the circle's radius)
- Circle area = πr²
- Ratio = (circle area) / (square area) = πr² / 4r² = π/4
- Therefore: **π ≈ 4 × (points inside circle) / (total points)**

This method is called "Monte Carlo" because it uses random sampling to solve a problem (named after the famous casino!).

```python
import random

# Method 2: Monte Carlo Simulation
def calculate_pi_monte_carlo(num_samples):
    """
    Estimate PI using the Monte Carlo method (random sampling).
    
    Args:
        num_samples: Number of random points to generate
        
    Returns:
        Estimated value of PI
    """
    # Count points that fall inside the circle
    points_inside_circle = 0
    
    # Generate random points and check if they're inside the circle
    for _ in range(num_samples):
        # Random point in the square [-1, 1] × [-1, 1]
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        
        # Distance from origin (0, 0)
        distance = math.sqrt(x**2 + y**2)
        
        # Check if point is inside the circle with radius 1
        if distance <= 1:
            points_inside_circle += 1
    
    # PI = 4 * (points inside circle) / (total points)
    pi_estimate = 4 * points_inside_circle / num_samples
    
    return pi_estimate

# Test with increasing number of samples
print("\nMonte Carlo Method (Random Sampling)")
print("=" * 50)
print(f"{'Samples':<15} {'PI Estimate':<20} {'Error':<15}")
print("-" * 50)

for samples in [1000, 10000, 100000, 1000000]:
    pi_est = calculate_pi_monte_carlo(samples)
    error = abs(pi_est - actual_pi)
    print(f"{samples:<15} {pi_est:<20.10f} {error:<15.10f}")

print(f"\nActual PI: {actual_pi:.10f}")
```

## Method 3: Infinite Series (Leibniz Formula)

### The Idea
Mathematicians discovered that PI can be calculated as an infinite series (an endless sum of numbers that approaches PI).

The **Leibniz formula** for PI is:

$$\frac{\pi}{4} = 1 - \frac{1}{3} + \frac{1}{5} - \frac{1}{7} + \frac{1}{9} - \frac{1}{11} + ...$$

Therefore:

$$\pi = 4 \times \left(1 - \frac{1}{3} + \frac{1}{5} - \frac{1}{7} + \frac{1}{9} - ...\right)$$

**How it works:**
1. Start with the first term (1)
2. Add and subtract fractions with odd denominators
3. The more terms you add, the closer you get to PI
4. This is a simple but slower method than the others!

```python
# Method 3: Leibniz Series
def calculate_pi_leibniz(num_terms):
    """
    Estimate PI using the Leibniz infinite series.
    
    Formula: π/4 = 1 - 1/3 + 1/5 - 1/7 + 1/9 - ...
    So:      π = 4 × (1 - 1/3 + 1/5 - 1/7 + 1/9 - ...)
    
    Args:
        num_terms: Number of terms in the series to calculate
        
    Returns:
        Estimated value of PI
    """
    leibniz_sum = 0
    
    # Calculate the series sum
    for n in range(num_terms):
        # Calculate the nth term
        # n=0: 1, n=1: 1/3, n=2: 1/5, n=3: 1/7, etc.
        denominator = 2 * n + 1
        
        # Alternate between adding and subtracting
        # n even: add, n odd: subtract
        if n % 2 == 0:
            leibniz_sum += 1 / denominator
        else:
            leibniz_sum -= 1 / denominator
    
    # Multiply by 4 to get PI
    pi_estimate = 4 * leibniz_sum
    
    return pi_estimate

# Test with increasing number of terms
print("\nLeibniz Series Method (Infinite Series)")
print("=" * 50)
print(f"{'Terms':<15} {'PI Estimate':<20} {'Error':<15}")
print("-" * 50)

for terms in [10, 100, 1000, 10000, 100000]:
    pi_est = calculate_pi_leibniz(terms)
    error = abs(pi_est - actual_pi)
    print(f"{terms:<15} {pi_est:<20.10f} {error:<15.10f}")

print(f"\nActual PI: {actual_pi:.10f}")
```

## Comparison of Methods

| Method | Speed | Accuracy | Difficulty | History |
|--------|-------|----------|------------|---------|
| **Polygon** | Fast | Very good | Medium | Used by Archimedes (287-212 BC) |
| **Monte Carlo** | Slow | Moderate | Easy | Modern probabilistic approach |
| **Leibniz Series** | Very Slow | Poor | Easy | Discovered in 1600s |

**Key Observations:**
- The **Polygon method** converges fastest - it reaches high accuracy with fewer calculations
- The **Monte Carlo method** is intuitive and practical for random simulations
- The **Leibniz series** is the simplest to understand mathematically but needs many terms
- All three methods prove the same fundamental truth: PI ≈ 3.14159...

## Appendix: Learning Jupyter Notebooks

### What is a Jupyter Notebook?
A Jupyter notebook is an interactive document that combines:
- **Markdown cells** - formatted text, explanations, and math
- **Code cells** - Python code you can run and see results
- **Output** - the results displayed below each code cell

### Key Features

**1. Cell Types**
- **Markdown cells** (like this one) - for explanations and documentation
- **Code cells** - contains Python code that you can execute

**2. Running Cells**
- Click the ▶ button or press `Ctrl+Enter` to run the selected cell
- Press `Shift+Enter` to run and move to the next cell
- Code cells execute in order, so earlier variables are available to later cells

**3. Cell Dependencies**
- Variables defined in earlier cells can be used in later cells
- For example, we used `actual_pi` (defined in the first code cell) in all three methods
- Always run cells from top to bottom!

**4. Advantages**
- Mix code, output, and explanations in one document
- Perfect for teaching and learning
- Great for data analysis and visualization
- Easy to share and collaborate
- Can be converted to HTML, PDF, or Python scripts

### Tips for Using Notebooks
- Run cells sequentially - don't skip around
- Use markdown cells to explain your thinking
- Add comments in your code (start with `#`)
- Re-run all cells if you restart the kernel
- Notebooks are `.ipynb` files - you can share them with others!
